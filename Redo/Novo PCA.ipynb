```python
import pandas as pd
```

```python
df = pd.read_csv ('meninasbin.csv', names = ['ID.Aluno',
 'Sexo',
 'Data.de.Nascimento',
 'UF.de.Nascimento',
 'Cotista',
 'Tipo.de.Escola',
 'Raca',
 'Curso',
 'Opçao',
 'Periodo.de.Ingresso.na.Unb',
 'Periodo.de.ingresso.na.opcao',
 'Forma.de.Ingresso.na.UnB',
 'Per\xc3\xadodo.de.saida.da.opcao',
 'Forma.de.saida.da.opcao',
 'Min..Cred..Formatura',
 'Creditos.cursados.no.total',
 'Creditos.integralizados.no.total'], sep=';')
```

```python
df = df.drop([0])
```

```python
df.reset_index()
```

```python
from sklearn.preprocessing import StandardScaler
features =  [
 'Cotista',
 'Tipo.de.Escola',
 'Raca',
 'Curso',
 'Opçao',
 'Periodo.de.Ingresso.na.Unb',
 'Periodo.de.ingresso.na.opcao',
 'Forma.de.Ingresso.na.UnB',
 'Per\xc3\xadodo.de.saida.da.opcao',
 'Min..Cred..Formatura',
 'Creditos.cursados.no.total',
 'Creditos.integralizados.no.total']
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['Forma.de.saida.da.opcao']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
finalDf = pd.concat([principalDf, df[['Forma.de.saida.da.opcao']].reset_index()], axis=1)
```

```python
finalDf = finalDf.drop(columns = 'index')
```

```python
finalDf
```

Se quiser saber como o PCA não normalizado fica, olhar no REDO Graficos

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['1', '2', '4']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Forma.de.saida.da.opcao'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
